### Stats Regarding Distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
 
import warnings
warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
mu, sigma = 1, 0.1
s = np.random.normal(mu, sigma, size = 10000)
print(len(s))
s
```

```python
plt.hist(s,100)
plt.show()
```

```python
sns.histplot(s, kde= True)
```

```python
# uniform distribution >> all outcomes has equal prob to occur in range
s = np.random.uniform(1,0, 10000)
sns.displot(s)
```

```python
# Logarithm Distribution >> whose logarithm is normlly ditributed.

l = np.random.lognormal(0,0.5, size = 10000)
sns.displot(l)
```

```python
plt.hist(l, bins=50)
```

```python
plt.hist(l[np.where(abs(l)<5)], bins=50)
```

```python
# log normal ---> normal
plt.hist(np.log(l), bins=50)
plt.show()
```

```python
sns.histplot(np.log(l), kde=True)
plt.show()
```

```python
#  sqaure to normal distribution
sns.histplot(np.sqrt(l), kde=True)
plt.show()
```

```python
#  chi square distribution
ch = np.random.chisquare(df=2,size=10000)
ch = ch[np.where(abs(ch)<10)]
plt.hist(ch, bins=50)
plt.show()
```

```python
sns.histplot(ch, kde=True)
```

```python
# /on increasing df, it behaves like normal distribution
ch = np.random.chisquare(df=50,size=10000)
# ch = ch[np.where(abs(ch)<5)]
sns.histplot(ch, kde=True)
```

```python
# t distribution
t = np.random.standard_t(df=1,size=1000)
t = t[np.where(abs(t)<10)]
sns.histplot(t, kde=True)
```

```python
t = np.random.standard_t(df=10,size=1000)
t = t[np.where(abs(t)<10)]
sns.histplot(t, kde=True)
```

* tail is bigger when DOF is 1 but more thinner/smaller when DOF =10

```python
# f distribution
# dfnum = numerator DOF, dfden = denominator DOF
f = np.random.f(dfnum=5,dfden = 8, size=10000)
f = f[np.where(abs(f)<10)]
sns.histplot(f, kde=True)
```

### Discrete Distribution

```python
# binomial Distribution >> n=1 for bernoulli distribution
b = np.random.binomial(n=1, p = 0.6, size=1000)
plt.hist(b)
```

```python
# binomial Distribution >> n>1 for binomial distribution
b = np.random.binomial(n=100, p = 0.6, size=1000)
plt.hist(b, bins = 50)
```

```python
# poisson ditribution
# lam = lambda,rate per hour
b = np.random.poisson(lam=5, size=1000)
plt.hist(b, bins=20)
```
