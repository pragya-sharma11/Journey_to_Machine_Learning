```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
x,y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=2, random_state=40)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)
```

### Grid Search CV >> Take all possible combinations

```python
from sklearn.model_selection import GridSearchCV
params = {'penalty': ('l1','l2','elasticnet'), 'C':[100,2,3,4,5,10,20,30,40]}
classifier  = LogisticRegression()
clf = GridSearchCV(estimator=classifier, param_grid=params, cv = 5, verbose=3)
```

```python
clf
```

```python
clf.fit(x_train, y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
LogisticRegression(penalty='l2', C = 1)
```

### Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
RandomizedSearchCV(classifier, param_distributions=params, cv = 5, verbose=3, n_iter=10)
clf.fit(x_train, y_train)
```

```python
clf.best_score_
```
