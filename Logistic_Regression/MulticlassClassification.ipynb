```python
from sklearn.datasets import load_iris
data = load_iris()
data.keys()
```

```python
print(data.DESCR)
```

```python
data.data
```

```python
data.target
```

```python
import pandas as pd
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
df.head()
```

```python
df['target'].unique()
```

```python
x = df[['sepal length (cm)','petal length (cm)']]
y = df['target'][:]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# there is a parameter called multiclass, which sets on auto, 
# if it is large category classes then use multinomial 
# if categories are lesser then use OVR
```

```python
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
model.predict_proba(x_test)
# give probablity for each 3 class
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, y_predict))
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import roc_curve, auc 
import matplotlib.pyplot as plt
y_pred_proba = model.predict_proba(x_test)
```
